## Experiment specific Functions

```python
PROPOSAL = 11019143
USER = getuser()
```

```python
BASEFOLDER = "/asap3/petra3/gpfs/p04/2023/data/%s/" % PROPOSAL
subfolder = "FGT"
sample_name = "2310_FGT"


# Load image files
def load_images(im_id):
    fname = join(BASEFOLDER, "raw", subfolder, "%s_%05d.nxs" % (sample_name, im_id))

    with nxload(fname) as f:
        im_out = np.array(f["scan"]["data"]["ccd"][()])
        print("Loaded: %s (%d frames)" % (fname, im_out.shape[0]))
    return im_out.squeeze()


# Load any kind of data from collection
def load_collection(scan_id, key):
    fname = join(BASEFOLDER, "raw", subfolder, "%s_%05d.nxs" % (sample_name, scan_id))

    with nxload(fname) as f:
        data = f["scan"]["instrument"]["collection"][key][()]
        print("Loaded: %s" % (fname))
    return data


# Load any kind of data from collection
def load_energy(scan_id):
    fname = join(BASEFOLDER, "raw", subfolder, "%s_%05d.nxs" % (sample_name, scan_id))

    with nxload(fname) as f:
        data = f["scan"]["instrument"]["mono"]["energy"][()]
        print("Loaded: %s" % (fname))
    return data


# Load any kind of data from measurements
def load_data(scan_id, key):
    fname = join(BASEFOLDER, "raw", subfolder, "%s_%05d.nxs" % (sample_name, scan_id))

    with nxload(fname) as f:
        data = f["scan"]["data"][key][()]
        print("Loaded key %s of %s" % (key, fname))
    return data


# Full image loading procedure
def load_processing(im_id):
    # Load data
    images = load_images(im_id)

    # Average over images
    if images.ndim > 1:
        image = np.mean(images, axis=0)
    else:
        image = images.copy()

    return image, images
```

# Experimental Details

```python
# Dict with most basic experimental parameter
experimental_setup = {
    "ccd_dist": 0.395,  # ccd to sample distance
    "energy": 707,  # x-ray energy
    "px_size": 15e-6,  # pixel_size of camera
    "binning": 1,  # Camera binning
}

# Convert eV to wavelength
experimental_setup["lambda"] = cci.photon_energy_wavelength(
    experimental_setup["energy"], input_unit="eV"
)

# Setup for azimuthal integrator
detector = Detector(
    experimental_setup["binning"] * experimental_setup["px_size"],
    experimental_setup["binning"] * experimental_setup["px_size"],
)

# General saving folder
folder_general = sup.create_folder(join(BASEFOLDER, "processed"))
print("Output Folder: %s" % folder_general)
```

# Load images


```python
# Specify image ids
im_ids = np.arange(966, 1045 + 1)
im_ids = [1378]

# Which data from nexus files to load? (e.g. "energy","m_magnett_read", ...)
key = "br1_master"

# Load data
images = []
scan_axis = []
for im_id in tqdm(im_ids):
    # load images
    image, timages = load_processing(im_id)

    if key == "energy":
        images.append(np.squeeze(image))
    else:
        images.append(np.squeeze(timages))

    # load data energy
    if key == "energy":
        scan_axis.append(np.squeeze(np.array(load_energy(im_id))))
    else:
        scan_axis.append(np.squeeze(np.array(load_data(im_id, key))))

images = np.squeeze(np.array(images))
scan_axis = np.squeeze(np.array(scan_axis))

# Plot scan axis
fig, ax = plt.subplots()
ax.plot(scan_axis, "o-")
ax.set_title("Scan Axis")
ax.set_ylabel(key)
ax.set_xlabel("Frame Index")
```

```python
# Assign to xarray
# Setup xarray for images
data = xr.Dataset()
data[key] = scan_axis
data["images"] = xr.DataArray(images, dims=[key, "y", "x"])

data["image"] = data["images"].mean(key)
image = data["image"].values

# Mask for integration over roi o..
mask = np.zeros(image.shape)
mask[1167:1331, 1100:1250] = 1

# Takes full image as roi
mask = np.ones(image.shape)

# Plot
mi, ma = np.nanpercentile(image, (1, 99))
fig, ax = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(8, 4))
ax[0].imshow(mask)
ax[0].set_title("Mask")
ax[1].imshow(mask)
ax[1].imshow(image, vmin=mi, vmax=ma, alpha=0.9)
ax[1].set_title("Avg Im Image with filled pixels")
```

```python
# Slideshow viewer
fig, ax = cimshow(data["images"])
```

## Basic widget to find center

Try to **align** the circles to the **center of the scattering pattern**. Care! Position of beamstop might be misleading and not represent the actual center of the hologram. 

```python
# Set center position via widget
ic = InteractiveCenter(image)
```

```python
# Get center positions
center = [ic.c0, ic.c1]
print(f"Center:", center)
shift_c = np.array(image.shape) / 2 - center
print(f"Shift vector:", shift_c)
```

## Azimuthal integrator widget for finetuning
If scattering pattern is radial symmetric, move center position until scattering ring is a line after transformation in polar coordinates

```python
# Setup azimuthal integrator for virtual geometry
ai = AzimuthalIntegrator(
    dist=experimental_setup["ccd_dist"],
    detector=detector,
    wavelength=experimental_setup["lambda"],
    poni1=center[0]
    * experimental_setup["px_size"]
    * experimental_setup["binning"],  # y (vertical)
    poni2=center[1]
    * experimental_setup["px_size"]
    * experimental_setup["binning"],  # x (horizontal)
)
```

```python
# Calc azimuthal integration
I_t, q_t, phi_t = ai.integrate2d(
    image,
    200,
    radial_range=(0, 0.06),
    unit="q_nm^-1",
    correctSolidAngle=False,
    dummy=np.nan,
    mask=1 - mask,
)
az2d = xr.DataArray(I_t, dims=("phi", "q"), coords={"q": q_t, "phi": phi_t})

# Plot
fig, ax = plt.subplots()
mi, ma = np.nanpercentile(I_t, [1, 98])
az2d.plot.imshow(ax=ax, vmin=mi, vmax=ma)
plt.title(f"Azimuthal integration")
```

```python
aic = AzimuthalIntegrationCenter(
    image,
    ai,
    c0=center[0],
    c1=center[1],
    im_data_range=[1, 99],
    radial_range=(0, 0.08),
    qlines=[40, 60],
)
```

```python
# Get center positions
center = [aic.c0, aic.c1]
print(f"Center:", center)
```

# Azimuthal integration for all images

```python
# Update center of azimuthal integrator
ai = AzimuthalIntegrator(
    dist=experimental_setup["ccd_dist"],
    detector=detector,
    wavelength=experimental_setup["lambda"],
    poni1=center[0]
    * experimental_setup["px_size"]
    * experimental_setup["binning"],  # y (vertical)
    poni2=center[1]
    * experimental_setup["px_size"]
    * experimental_setup["binning"],  # x (horizontal)
)
```

```python
# Do 2d Azimuthal integration of all images and append them to list
list_i2d = []
for i in tqdm(range(data.images.shape[0])):
    if key == "energy":
        ai.wavelength = cci.photon_energy_wavelength(
            data["energy"][i].values, input_unit="eV"
        )
    elif key == "br1_master":
        ai.wavelength = cci.photon_energy_wavelength(
            data[key][i].values, input_unit="eV"
        )

    im = data["images"][i].values
    i2d, q, chi = ai.integrate2d(im, 500, 90, dummy=np.nan, mask=1 - mask)
    list_i2d.append(i2d)

data["q"] = q
data["chi"] = chi
data["i2d"] = xr.DataArray(list_i2d, dims=[key, "chi", "q"])
```

```python
# Plot
fig, ax = plt.subplots()
mi, ma = np.percentile(I_t, [0.1, 99])
data["i2d"][7].plot.imshow(ax=ax, vmin=610, vmax=630)
plt.title(f"Azimuthal integration")
```

```python
# Select chi-range
sel = (data.chi > -40) * (data.chi < 125) + (data.chi < -76) * (data.chi > 100)
data["i1d"] = data.i2d.where(sel, drop=True).mean("chi")
# data["i1d"] = data.i2d.mean("chi")
```

```python
# Plot 1d azimuthal integration of frame idx to estimate the relevant q-range
idx = 0
fig, ax = plt.subplots()
ax.plot(data.q, data.i1d[idx] + 1)
ax.set_yscale("log")
ax.set_title("Frame Index: %d" % idx)
ax.set_xlabel("q")
ax.set_ylabel("Azimuthal Integration")
```

```python
# Select relevant q-range for averaging
q0, q1 = 0.012, 0.03
sel = (data.q > q0) * (data.q < q1)
data["saxs"] = data.i1d.where(sel, drop=True).mean("q")

# Plot Intensity of SAXS Pattern
fig, ax = plt.subplots()
ax.plot(data[key].values, data["saxs"].values)
ax.grid()
ax.set_xlabel(key)
ax.set_ylabel("Integrated SAXS")

# Title and fname
try:
    ax.set_title("Scan Id %s-%s" % (im_ids[0], im_ids[-1]))
    fname = "ImId_%04d-%04d_%s.png" % (im_ids[0], im_ids[-1], USER)
except:
    ax.set_title("Scan Id %d" % (im_ids))
    fname = "ImId_%04d_%s.png" % (im_ids, USER)

print("Saving:%s" % fname)
plt.savefig(path.join(folder_general, fname))
```
